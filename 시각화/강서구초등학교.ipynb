```python
from folium.map import Popup
import pandas as pd
import folium

df=pd.read_csv('서울특별시강서구학교기본정보.csv')
seoul_map=folium.Map(location=[37.55,126.88],titles='Stamen Terrain',
                    zoom_start=13)
#위치 정보를 Marker 로 표시
#index를 학교명으로 변경하고 시설 이름이 표시되도록 함
for name, lat, lng in zip(df.학교명,df.위도,df.경도):
    iframe=folium.IFrame(name,width=100,height=80)
    popup=folium.Popup(iframe,max_width=500)
    folium.Marker([lat,lng],popup=popup).add_to(seoul_map)
    
seoul_map

#초등학교 밀집도 분석
#지역 아동 센터 비율과 분석 필요
#등촌동 6개, 화곡동 7개의 초등학교 입지 
#등촌동, 화곡동에 입지하고 있는 아동시설에 다문화 가정 학생들을 위한 멘토링 시범 운영
#열린지역지역아동센터, 등촌동에는 위치하고 있는 아동 시설이 없음
#염창동에 있는 한길지역 아동센터 연결
```
